```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# Read census csv file

census_file = pd.read_csv('census_data.csv')
census_file
```

```python
# Clean up the cencus dataframe

census_pd = pd.DataFrame(census_file)

census_pd[census_pd <= 0] = np.nan
census_pd.dropna(axis=0, how='any',inplace=True)
census_pd
```

```python
census_pd
```

```python
census_pd.head()
```

```python
# Select columns interested
slim_cencus_pd = census_pd[['Zipcode', 'Population', 'Median Age','Household Income']]
slim_cencus_pd = slim_cencus_pd.rename(columns={'Median Age':'Median_Age',
                                    'Household Income':'Household_Income'
                                   })

slim_cencus_pd.to_csv("cleaned_census_data.csv",index=False)
slim_cencus_pd.head()
```
